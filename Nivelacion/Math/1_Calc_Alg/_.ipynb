## 1. Conjuntos numéricos & operaciones (page 1)
- ℕ ⊂ ℤ ⊂ ℚ ⊂ ℝ ⊂ ℂ
- **Propiedades**
  - Conmutativa: $a+b=b+a, a·b=b·a$
  - Asociativa: $(a+b)+c=a+(b+c), (a·b)·c=a·(b·c)$
  - Distributiva: $a(b+c)=ab+ac$
  - Neutros: $a+0=a, a·1=a$
  - Inverso: $a+(-a)=0, a·a⁻¹=1 (a≠0)$
- **Potencias**
  $aⁿ aᵐ = aⁿ⁺ᵐ, (aⁿ)ᵐ = aⁿᵐ, a⁻ⁿ = \frac{1}{aⁿ}$
- **Radicales**
  $\sqrt[n]{a}$ = a¹ᐟⁿ; racionalizar multiplicando por el conjugado.

## 2. Expresiones algebraicas (page 17)
- **Monomio**: $a xᵏ$ (grado k, coeficiente a)
- **Polinomio**: $P(x)=Σ aᵢxⁱ, gr(P)=máxGradoCona≠0$
- **Valor numérico**: $P(x)=Σ aᵢ xⁱ$
- **Operaciones** (page 24)
  - Suma/resta: agrupar términos del mismo grado.
  - Multiplicación: distributiva; $gr(P·Q)=gr(P)+gr(Q)$.
  - División: algoritmo de Ruffini o coeficientes; resto $R=P(α)$.
- **Identidades notables**
  $(a±b)² = a² ± 2ab + b²$
  $(a+b)(a−b)=a²−b²$
  $(a±b)³ = a³ ± 3a²b + 3ab² ± b³$

## 3. Factorización (page 45)
- Factor común: $ax+ay=a(x+y)$
- Diferencia de cuadrados: $a²−b²=(a−b)(a+b)$
- Trinomio cuadrado perfecto: $a²±2ab+b²=(a±b)²$
- Suma/resta de cubos: $a³±b³=(a±b)(a²∓ab+b²)$
- Ruffini para raíces racionales: divisores del término independiente.

## 4. Fracciones algebraicas (page 54)
- Simplificar factorizando numerador y denominador.
- Suma/resta: MCM de denominadores.
- Multiplicación/ división: simplificar antes de operar.

## 5. Ecuaciones lineales (page 37)
- Forma $ax+b=0 ⇒ x=\frac{−b}{a}$
- Problemas: plantear igualdad → despejar incógnita.

## 6. Sistemas 2×2 (page 33)
- **Métodos**
  - Sustitución: despejar una incógnita y reemplazar.
  - Igualación: despejar la misma incógnita en ambas y igualar.
  - Reducción: multiplicar para igualar coeficientes y sumar/restar.
- **Discusión**
  - Determinado: solución única.
  - Incompatible: sin solución.
  - Indeterminado: infinitas soluciones.

## 7. Ecuaciones cuadráticas (page 45)
- Forma estándar: $ax²+bx+c=0 (a≠0)$
- **Discriminante**: $Δ=b²−4ac$
  - $Δ>0$: 2 raíces reales distintas.
  - $Δ=0$: raíz doble $x=\frac{−b}{(2a)}$.
  - $Δ<0$: 2 raíces complejas conjugadas.
- **Fórmula resolutiva**:
  $x = \frac{−b ± \sqrt{Δ}}{2a}$
- **Relaciones de Vieta**:
  $x₁+x₂ = \frac{−b}{a}, x₁x₂ = \frac{c}{a}$
- **Completar cuadrados**:
  $x²+px = (x+\frac{p}{2})² − (\frac{p}{2})²$

## 8. Números complejos (page 14)
- Forma binómica: $z=a+bi$
- Conjugado: $z̄=a−bi$
- **Operaciones**
  - Suma/resta: $(a+bi)±(c+di)=(a±c)+(b±d)i$
  - Producto: $(a+bi)(c+di)=(ac−bd)+(ad+bc)i$
  - Inverso: $\frac{1}{z} = \frac{z̄}{a²+b²}$
- **Propiedades clave**: $z+z̄=2a∈ℝ, z·z̄=a²+b²≥0$

## 9. Casos particulares frecuentes
- Raíces opuestas: $x²−k=0 ⇒ x=± \sqrt{k}$
- Raíces complejas conjugadas: if una es $p+qi$, la otra es $p−qi$
- Raíz doble: $Δ=0 ⇒ x=\frac{−b}{2a}$
- Coeficientes enteros: usar Ruffini para raíces racionales.
- Parámetros: imponer $Δ=0$ o usar Vieta para hallar el parámetro.

## 10. Estrategia de resolución universal
1. Identificar tipo (lineal, cuadrática, sistema, fracción).
2. Factorizar/simplificar if es posible.
3. Aplicar método específico (despeje, fórmula, sustitución, etc.).
4. Verificar soluciones en la ecuación original.
5. Interpretar en el contexto del problema (if aplica).
